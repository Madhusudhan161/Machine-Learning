### 1. What is the difference between ordinal encoding and label encoding ? provide an example of when you might choose one over the other .

|                  Ordinal encoding        |               Label Encoding         |
|-----------------------------------------|---------------------------------------|
|1. ordinal encoding assigns a unique integer value to each category of a categorical variable based on order or hierarchy | 1. Label encoding assigns a unique integer value to each category of categorical varibale without considering any order or hierarchy |
|2. it preserves the ordinal relationship between the categories assuming that there is a meaningful order or rank|2. it does not preserve any ordinal relationship between the categories and treats them as disinct and unrelated | 
|3. ordinal encoding is suitable for ordinal categorical varibles where the categories have a natural order or hierarchy | 3. Label encoding is suitable for nominal categorical variables where the categories have no inherent order or hierarchy | 

Example : 
- consider a dataset containing a categorical variable representing customer satisfaction with 'poor','fair','good','excellent'.

1. Ordinal encoding : if the satisfaction levels have natural order or hierarchy (poor<fair<good<excellent) we can use ordinal encoding to preserve this order.

2. Label encoding : if the satisfaction levels do not have meaningful order or hierarchy are treated as distinct categories with no inherent ranking. 

### 2. Explain how Target Guided Ordinal Encoding works and provide an example of when you might use it in a machine learning project.

- Target guided ordinal encoding is a technique used to transform categorical varibales into ordinal variables based on target varibales relationship with the categories. 

##### how it works :
1. Calculate mean/median of target variable : for each category of categorical variable, compute the mean or median value of the target varibale. 
2. Rank categories : sort the categories based on their mean or median target varibale values. lower ranks given to lower mean vice versa.
3. Replace categories with ranks : Replace the original categorical values with thier corresponding ranks. 

Example scenario to use target guided ordinal encoding : 

- A project to predict customer default risk for a bank. One of the categorical features in dataset is "employment_status", which indicates the employment status of customers (e.g., employed, self-employed, unemployed, retired). encode this categorical variable into ordinal format based on its relationship with the target variable, which is whether the customer defaults on their loan or not.

usage : 
1. Calculate mean or median default rate for each category of 'employment_status'
2. rank the categories based on thier mean or meadian default rates, with lower ranks assigned to categories with lower default rates..etc
3. Replace orginal categorical values with thier corresponding ranks
4. use the encoded ordinal variable in your machine learning model to predict customer default risk. 

### 3. Define covariance and explain why it is important in statistical analysis. How is covariance calculated?

- Covariance is a statistical measure that quantifies the degree to which two random variables change together. In other words, it measures the relationship between two variables and indicates whether they move in the same direction (positive covariance), opposite directions (negative covariance), or are independent (zero covariance)

### 4. For a dataset with the following categorical variables: Color (red, green, blue), Size (small, medium, large), and Material (wood, metal, plastic), perform label encoding using Python's scikit-learn library. Show your code and explain the output.

```python
## importing modules 
import pandas as pd
from sklearn.preprocessing import LabelEncoder

## given data 
df = pd.DataFrame({
    'color':['red','green','blue'],
    'size':['small','medium','large'],
    'material':['wood','metal','plastic']
})

## initiating the encoder 
label = LabelEncoder()

## using fit and transform on data
label.fit_transform(df[['color']])
```

```python
label.fit_transform(df[['size']])
```

```python
label.fit_transform(df[['material']])
```

### 5. Calculate the covariance matrix for the following variables in a dataset: Age, Income, and Education level. Interpret the results.

```python
## importing required modules
import numpy as np
import pandas as pd

## Given data
df = pd.DataFrame({
    'Age': [30, 40, 35, 45],
    'Income': [50000, 60000, 55000, 65000],
    'Education level': [1, 2, 3, 2]  # ordinal data i.e highschool = 1, bachelors = 2..etc
})

## Calculate covariance matrix
covariance_matrix = df.cov()

## printing the results
print("Covariance Matrix:")
print(covariance_matrix)
```

### 7. You are analyzing a dataset with two continuous variables, "Temperature" and "Humidity", and two categorical variables, "Weather Condition" (Sunny/Cloudy/Rainy) and "Wind Direction" (North/South/ East/West). Calculate the covariance between each pair of variables and interpret the results.

```python
## importing modules
import numpy as np
import pandas as pd

## given data 
temperature = [25,28,30,22,27]
humidity = [60,55,70,45,50]

## calculating covariance between humidity and temperature
cov_hum_temp = np.cov(temperature,humidity)[0,1]

## printing the results 
print('Covariance between temperature and Humidity:',cov_hum_temp)
```

##### note : we can calculate covariance for continous data, but cannot do it with continous and categorical data together. as covariance is a measure of two variables change together.
